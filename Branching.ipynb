ITERATION USING OF LOOPS -----> WHILE AND FOR 

```python
%%time

result = 1
i = 1

while i <= 1000:
    result *= i # same as result = result * i
    i += 1 # same as i = i+1

print(result)
```

```python
result = 1
i = 1

while i <= 100:
    result = result * i
    i = i+1

print('The factorial of 100 is: {}'.format(result))
```

```python
line = '*'
max_length = 10

while len(line) <= max_length:
    print(line)
    line += "*"
    
while len(line) > 0:
    print(line)
    line = line[:-1]
```

```python
line = '*'
max_length = 10

while len(line) <= max_length:
    print(line)
    line += "*"
while len(line) > 0:
    print(line)
    line = line[:-1]    
```

```python
line = "*"
limit = 6
i = 1
while len(line) <= limit:
    print(line * len(line))
    line = line + len(line)
    
```

```python
i = 1
result = 1

while i < 20:
    i += 1
    if i % 2 == 0:
        print('Skipping {}'.format(i))
        continue
    print('Multiplying with {}'.format(i))
    result = result * i
    
print('i:', i)
print('result:', result)
```
